```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv('Mall_Customers.csv')

# 기본 정보 확인
print(df.head())         # 데이터 상위 5행 확인
print(df.info())         # 데이터 타입과 결측치 확인
print(df.describe())     # 수치형 데이터의 요약 통계량
```

```python
# 결측값 확인
print(df.isnull().sum())
```

```python
# 데이터 타입 확인
print(df.dtypes)
```

```python
# 중복값 확인
print(df.duplicated().sum())
```

```python
# sns.boxplot(x = df['Age'])
# plt.show()

# sns.boxplot(x = df['Gender'])
# plt.show()

sns.boxplot(x = df['Annual Income (k$)'])
plt.show()

# sns.boxplot(x = df['Spending Score (1-100)'])
# plt.show()
```

```python
# 중복값 확인
print(df.duplicated().sum())
print(df.duplicated())
```

```python
# 평균
print('평균 : ',df['Annual Income (k$)'].mean())
# 중앙값
print('중앙값 : ', df['Annual Income (k$)'].median())
# 최빈값
print('최빈값 : ', df['Annual Income (k$)'].mode()[0])
# 분산값
print('분산 값 : ', df['Annual Income (k$)'].var())
# 표준편차
print('표준편차 : ', df['Annual Income (k$)'].std())
# 최고값,최소값
print(f'최고값 : {df['Annual Income (k$)'].max()}, 최소값 : {df['Annual Income (k$)'].min()}')
```

```python
from scipy.stats import skew, kurtosis

# 왜도 첨도 확인
print('왜도 : ', skew(df['Annual Income (k$)']))
print('첨도 : ', kurtosis(df['Annual Income (k$)']))
```

```python
# 페어플롯 생성
sns.pairplot(df)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.suptitle('구매고객 데이터 변수간 관계', y=1.02)
plt.show()
```
